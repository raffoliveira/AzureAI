```python
import requests
from bs4 import BeautifulSoup
from langchain_openai.chat_models.azure import AzureChatOpenAI

subscription_key = "key"
```

```python
from os import sep
def extract_text_from_url(url):
  response = requests.get(url)
  if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    for script in soup(["script", "style"]):
      script.decompose()
    text = soup.get_text(separator= " ")
    lines = (line.strip() for line in text.splitlines())
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    text = '\n'.join(chunk for chunk in chunks if chunk)
    return text
  else:
    return None
```

```python
extract_text_from_url(url="https://dev.to/rofiatabdulkareem/azure-fundamentals-4i0h")
```

```python
client = client = AzureChatOpenAI(
    api_version="2024-12-01-preview",
    azure_endpoint="https://openairafa.openai.azure.com/",
    api_key=subscription_key,
    deployment_name="gpt-4o-mini",
    max_retries=0
)
```

```python
def translate_article(text, lang):
  messages = [
      ("system", "You act as text translator"),
      ("user", f"Translate the following text to {lang}:\n{text} and answer in markdown")
  ]
  response = client.invoke(messages)
  return response.content
```

```python
translate_article("An Azure region is a geographical area containing one or more data centers, which are interconnected through a low-latency network.",
                  "portugues")
```

```python
article = extract_text_from_url(url="https://dev.to/rofiatabdulkareem/azure-fundamentals-4i0h")
translate_article(article, "portugues")
```
